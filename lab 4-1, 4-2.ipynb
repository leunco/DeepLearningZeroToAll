## 4.1 Multivariable linear regresssion

> 지난 시간  

**linear regression을 설계하기 위해 필요한 조건(입력이 하나일 때)**
1. Hypothesis  
$$H(x)=Wx+b$$


$$W : weight, b : bias$$


2. Cost function  
$$cost(W,b) = {1\over m}\sum_{i=1}^m(H(x^{(i)})-y^{(i)})^2$$  
$$예측값 - 실제값(true)$$


3. Gradient descent algorithm  
  - cost 최적화 값 찾기   
<br>

---

**Multivariable linear regression**
  - 입력값이 3개 일 때  

1. Hypothesis
$$H(x_1,x_2,x_3) = w_1x_1+w_2x_2+w_3x_3+b$$


2. Cost function
$$cost(W,b) = {1\over m}\sum_{i=1}^m(H(x_1^{(i)},x_2^{(i)},x_3^{(i)})-y^{(i)})^2$$  


**Hypothesis using matrix**

$$w_1x_1+w_2x_2+w_3x_3+...+w_nx_n$$  
$$\left[ \begin{array}{ccc}
x_{11} & x_{12} & x_{13} \\
x_{21} & x_{22} & x_{23} \\
x_{31} & x_{32} & x_{33} \\
x_{41} & x_{42} & x_{43} \\
x_{51} & x_{52} & x_{53} \\ \end{array} \right]  
*  \left[ \begin{array}{c}
        w_{1} \\
        w_{2} \\
        w_{3} \\ \end{array} \right] = \left[ \begin{array}{c}
x_{11}w_1 + x_{12}w_2 + x_{13}w_3 \\
x_{21}w_1 + x_{22}w_2 + x_{23}w_3 \\
x_{31}w_1 + x_{32}w_2 + x_{33}w_3 \\
x_{41}w_1 + x_{42}w_2 + x_{43}w_3 \\
x_{51}w_1 + x_{52}w_2 + x_{53}w_3 \\ \end{array} \right]$$  

<br>
<br>
$$H(x) = XW$$

- [5,3] * [3,1] = [5,1]  
- <u>[5,3]에서 5는 인스턴스 개수, 3은 variable 개수</u>


- [n,**3**] * [**3**,**1**] = [n,**1**]  
    - n : 여러 개 들어올 수 있음(넘파이에서는 -1, 텐서플로에서는 None으로 표현)  


**WX vs XW **  
1. Lecture(theory)   
$$H(x) = Wx+b$$


2. Implementation(TensorFlow)  
$$H(x)=XW$$

## lab 4.1 multi-variable linear regression 구현

**lab-04-1-multi_variable_linear_regression**

```python
import tensorflow as tf

tf.set_random_seed(777) # for reproducibility

x1_data = [73., 93., 89., 96., 73.]
x2_data = [80., 88., 91., 98., 66.]
x3_data = [75., 93., 90., 100., 70.]

y_data = [152., 185., 180., 196., 142.]

# placeholders for a tensor that will be always fed
x1 = tf.placeholder(tf.float32)
x2 = tf.placeholder(tf.float32)
x3 = tf.placeholder(tf.float32)
Y = tf.placeholder(tf.float32)

# weight과 bias는 하나의 값 -> [1]
w1 = tf.Variable(tf.random_normal([1]),name='weight1') # 변수 생성
w2 = tf.Variable(tf.random_normal([1]),name='wegith2')
w3 = tf.Variable(tf.random_normal([1]),name='weight3')
b = tf.Variable(tf.random_normal([1]),name='bias')
```

```python
hypothesis = x1 * w1 + x2 * w2 + x3 * w3 + b # H(x1,x2,x3) = x1w1+x2w2+x3w3

# cost/loss function
cost = tf.reduce_mean(tf.square(hypothesis - Y))

# Minimize: Need a very small learning rate for this data set
optimizer = tf.train.GradientDescentOptimizer(learning_rate=1e-5)
train = optimizer.minimize(cost)

# Launch the graph in a session.
sess = tf.Session()

# Initializes global variables in the graph.
sess.run(tf.global_variables_initializer())

# 학습
for step in range(2001):
    cost_val, hy_val, _ = sess.run([cost, hypothesis, train],
                                  feed_dict={x1: x1_data, x2:x2_data, x3:x3_data, Y:y_data})
    
    if step % 10 == 0:
        print(step, "Cost: ", cost_val, "\nPredict:\n", hy_val)
```

>행렬 이용

**lab-04-2-multi_variable_linear_regression**

$$\left[ \begin{array}{ccc}
x_{1} & x_{2} & x_{3} \\ \end{array} \right]  
*  \left[ \begin{array}{c}
        w_{1} \\
        w_{2} \\
        w_{3} \\ \end{array} \right] = \left[ \begin{array}{c}
x_{1}w_1 + x_{2}w_2 + x_{3}w_3 \\ \end{array} \right]$$  

<br>
<br>
$$H(x) = XW$$

```python
tf.set_random_seed(777)

x_data = [[73., 80., 75.],
         [93., 88., 93.],
         [89., 91., 90.],
         [96., 98., 100.],
         [73., 66., 70.]]

y_data = [[152.],
         [185.],
         [180.],
         [196.],
         [142.]]

### 여기 행렬 이용해서 간단하게 구현 ###
# placeholders for a tensor that will be always fed
X = tf.placeholder(tf.float32, shape=[None,3]) # x1,x2,x3(각 element) 3개라서 3, 앞에 None은 필요에 따라 x를 추가할 수 있기 때문에 n개의 데이터를 가지고 있음을 의미
Y = tf.placeholder(tf.float32, shape=[None,1])
W = tf.Variable(tf.random_normal([3,1]),name='weight')
b = tf.Variable(tf.random_normal([1]),name='bias')

# Hypothesis
hypothesis = tf.matmul(X,W)+b

# Simplified cost/loss function
cost = tf.reduce_mean(tf.square(hypothesis-Y))

# Minimize
optimizer = tf.train.GradientDescentOptimizer(learning_rate=1e-5)
train = optimizer.minimize(cost)

# Launch the graph in a session
sess = tf.Session()

# Initialies global variables in the grpah
sess.run(tf.global_variables_initializer())

for step in range(2001):
    cost_val, hy_val, _ = sess.run(
    [cost, hypothesis, train], feed_dict={X: x_data, Y: y_data})
    if step % 10 == 0:
        print(step, "Cost:", cost_val, "\nPrediction:\n", hy_val)
```

```python
import tensorflow as tf
tf.set_random_seed(777)  # for reproducibility

filename_queue = tf.train.string_input_producer(
    ['data-01-test-score.csv'], shuffle=False, name='filename_queue')

reader = tf.TextLineReader()
key, value = reader.read(filename_queue)

# Default values, in case of empty columns. Also specifies the type of the
# decoded result.
record_defaults = [[0.], [0.], [0.], [0.]]
xy = tf.decode_csv(value, record_defaults=record_defaults)

# collect batches of csv in
train_x_batch, train_y_batch = \
    tf.train.batch([xy[0:-1], xy[-1:]], batch_size=10) 
                  # x data     y data
# tf.train.batch 함수가 배치로 묶어서 리턴을 하는 함수인데, batch로 묶고자 하는 tensor 들을 인자로 준 다음에, 
# batch_size (한번에 묶어서 리턴하고자 하는 텐서들의 개수)를 정해주면 된다.
# 출처: https://bcho.tistory.com/1170 [조대협의 블로그]
                 

# placeholders for a tensor that will be always fed.
X = tf.placeholder(tf.float32, shape=[None, 3])
Y = tf.placeholder(tf.float32, shape=[None, 1])

W = tf.Variable(tf.random_normal([3, 1]), name='weight')
b = tf.Variable(tf.random_normal([1]), name='bias')

# Hypothesis
hypothesis = tf.matmul(X, W) + b

# Simplified cost/loss function
cost = tf.reduce_mean(tf.square(hypothesis - Y))

# Minimize
optimizer = tf.train.GradientDescentOptimizer(learning_rate=1e-5)
train = optimizer.minimize(cost)

# Launch the graph in a session.
sess = tf.Session()
# Initializes global variables in the graph.
sess.run(tf.global_variables_initializer())

# Start populating the filename queue.
coord = tf.train.Coordinator()
threads = tf.train.start_queue_runners(sess=sess, coord=coord)

for step in range(2001):
    x_batch, y_batch = sess.run([train_x_batch, train_y_batch])
    cost_val, hy_val, _ = sess.run(
        [cost, hypothesis, train], feed_dict={X: x_batch, Y: y_batch})
    if step % 10 == 0:
        print(step, "Cost: ", cost_val, "\nPrediction:\n", hy_val)

coord.request_stop()
coord.join(threads)
```

```python
# Ask my score
print("Your score will be ",
      sess.run(hypothesis, feed_dict={X: [[100, 70, 101]]}))

print("Other scores will be ",
      sess.run(hypothesis, feed_dict={X: [[60, 70, 110], [90, 100, 80]]}))
```
